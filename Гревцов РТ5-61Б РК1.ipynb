```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("HousingData.csv")
print(data.head())
```

```python
# Обработка пропусков в данных
data_cleaned = data.dropna()
```

```python

# Корреляционный анализ
corr = data_cleaned.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Корреляционная матрица')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```

```python
# Создание скрипичной диаграммы для произвольной колонки данных
plt.figure(figsize=(8, 6))
sns.violinplot(y=data_cleaned['AGE'], color='skyblue', linewidth=2)
plt.title('Скрипичная диаграмма для возраста зданий (AGE)')
plt.xlabel('Возраст зданий')
plt.ylabel('Плотность')
plt.grid(True)
plt.show() 
```

## Выводы

По матрице корреляции сделаем выводы о взаимосвязи между признаками.

- **RM** имеет сильную положительную корреляцию с ценой жилья. Следовательно, **RM** является очень важным признаком для предсказания цены.
- **LSTAT** (процент населения с низким статусом) имеет сильную отрицательную корреляцию, означающую, что более высокий процент бедного населения в районе обычно связан с более низкими ценами на жилье.
- **NOX** (концентрация оксидов азота) имеет отрицательную корреляцию, указывая на то, что увеличение загрязнения связано с снижением стоимости жилья.

Сильная взаимосвязь признаков с целевой переменной может указывать на их полезность для моделей машинного обучения.

---

Признаки с высокой корреляцией (как положительной, так и отрицательной) с целевой переменной могут считаться важными предикторами. Важно учитывать и мультиколлинеарность, которая может негативно повлиять на производительность моделей машинного обучения.

---

Для задачи регрессии, как в данном случае предсказания цены на недвижимость, подходящими моделями будут линейная регрессия, решающие деревья, случайный лес и градиентный бустинг. Выбор оптимальной модели будет зависеть от сложности данных, требуемой точности и интерпретируемости модели. В данном случае классической модели линейной регрессии будет достаточно.
